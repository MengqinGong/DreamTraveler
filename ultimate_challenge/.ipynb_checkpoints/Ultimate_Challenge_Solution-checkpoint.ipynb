```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
```

### Part I Exploratory data analysis

```python
# Load login.json
logins = pd.read_json('logins.json')
print(logins.head())
```

```python
# Convert to timestamp
logins.login_time = pd.to_datetime(logins.login_time, format='%Y-%m-%d %H:%M:%S')
print(type(logins.login_time[0]))
```

```python
# Group records based on 15min time intervals
logins_count = logins.resample('15T', on='login_time').count()
login_agg = pd.DataFrame(index=logins_count.index, data=logins_count.values, columns=['count'])
```

```python
# Calculate total number of rides each day
daily_ride = login_agg.resample('D').sum()

plt.figure(figsize=(10,6))
plt.plot(daily_ride)
plt.xlabel('Date')
plt.ylabel('Number of Logins')
plt.show()
```

```python
# Calculate average rides by hour of the day
hourly_ride = round(login_agg.groupby(login_agg.index.hour).mean(),0)

plt.figure(figsize=(10,6))
plt.bar(hourly_ride.index, hourly_ride['count'])
plt.xlabel('Hour')
plt.ylabel('Average Number of Logins')
plt.show()
```

```python
# Calculate average rides by day of the week
weekday_ride = round(login_agg.groupby(login_agg.index.weekday).mean(),0)
weekday = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
plt.figure(figsize=(10,6))
plt.bar(weekday_ride.index, weekday_ride['count'], tick_label=weekday)
plt.xlabel('Weekday')
plt.ylabel('Average Number of Logins')
plt.show()
```

### Part III Predictive Modeling

```python
import json
with open('ultimate_data_challenge.json') as datafile:
    data = json.load(datafile)
users = pd.DataFrame(data)
users.head()
```

```python
users.info()
```

According to the dataframe info, we can see that we have 2 categorical variables: city and phone; 2 datetime variables: last_trip_date and signup_date; 1 boolean variable: ultimate_black_user and others are numerical variables.<br>
<br>
We have missing data in the following 3 variables: avg_rating_by_driver, avg_rating_of_driver and phone.<br>
<br>
First let's deal with missing data.

```python
users.phone.value_counts()
```

For 'phone', there are only two values: 'iPhone' and 'Android'. It is unlikely that the value is missing at random, possible reason can be the phone is not a smart phone (if Ultimate can be run on a non-smart phone) or the operation system is unknown for our data collection process. I fill missing values with 'NA'.<br>
<br>
For avg_rating_by_driver and avg_rating_of_driver, the possible reason for missing value can be the user is new to Ultimate and has never rated or been rated yet, or the user/driver just forgot to rate after the ride. Therefore, I will fill the missing value with the median so that they will not affect the median of the 2 features.

```python
fill_values = {'phone': 'NA', 
               'avg_rating_by_driver': users.avg_rating_by_driver.median(), 
               'avg_rating_of_driver': users.avg_rating_of_driver.median()}
users = users.fillna(value=fill_values)
```

```python
users.last_trip_date.max()
```

Since the last day of last_trip_date is 2014-07-01, we assume the data was pulled on that day, and an active user should be an user who had a trip within 30 days before 2014-07-01, that is, 2014-06-01.

```python
# create a Active column: True if last_trip_date > '2014-06-01', otherwise False
users['active'] = (users.last_trip_date >= '2014-06-01')
users.head()
```

```python
# calculate the retention rate
print(users.active.value_counts())
retention_rate = users.active.value_counts()[True]/users.shape[0]
print(retention_rate)
```

##### Exploratory Data Analysis

First let's check out the relationship between numeric variables and retention

```python
numeric_cols = ['avg_dist', 'avg_rating_by_driver', 'avg_rating_of_driver', 'avg_surge',
                'trips_in_first_30_days', 'surge_pct', 'weekday_pct', 'days_since_signup']

plt.figure(figsize=(16,16))
for i in range(len(numeric_cols)):
    plt.subplot(3,3,i+1)
    sns.boxplot(x='active', y=numeric_cols[i], data=users, showfliers=False)   
```

For categorical variables, I perform Chi-squared test to test the independency between them.

```python
# active vs. phone
from scipy.stats import chi2_contingency

crosstab_active_phone = pd.crosstab(users.active, users.phone)
crosstab_active_phone
```

```python
chi2_contingency(crosstab_active_phone)
```

The first value (2563.03) is the Chi-square value, followed by the p-value (0.0), then comes the degrees of freedom (2), and lastly it outputs the expected frequencies as an array. <br>
<br>
In this case we have evidence to reject the null hypothesis that 'phone' and 'active' are independent to each other. That is, phone variable is related to activeness/retention. 

Now let's deal with categorical and datetime variables.<br>
<br>
Categorical variables: create dummy variables<br>
<br>
Datetime: calculate the number of the date to pull day, and drop the two datetime variables.

```python
# active vs. city
crosstab_active_city = pd.crosstab(users.active, users.city)
crosstab_active_city
```

```python
chi2_contingency(crosstab_active_city)
```

The p-value of (0.0) suggests that we need to reject the null hypothesis, that means the test shows that active is related to city.

```python
# active vs. ultimate_black_user
crosstab_active_ultimate_black_user = pd.crosstab(users.active, users.ultimate_black_user)
crosstab_active_ultimate_black_user
```

```python
chi2_contingency(crosstab_active_ultimate_black_user)
```

The p value is 0.0 << 0.05, reject null hypothesis. Conclusion: active and ultimate_black_user are not independent to each other.<br>
<br> 
Now let's do some further data wrangling.

```python
# convert datetime columns to datetime type
users.last_trip_date = pd.to_datetime(users.last_trip_date)
users.signup_date = pd.to_datetime(users.signup_date)

# calculate the day difference between last day of last_trip_date and signup_date 
pull_date = pd.to_datetime(users.last_trip_date.max())
users['days_since_signup'] = (pull_date - users.signup_date).dt.days

# drop last_trip_date and signup_date
users = users.drop(['last_trip_date', 'signup_date'], 1)

# create dummies for phone and city
users = pd.get_dummies(users)

users.head()
```

```python
# create model input X, y
y = users.active
X = users.drop('active', 1)

# split X, y into train set(70%) and test set(30%)
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=66)

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

This is a classification problem. Random Forest Classifier is a powerful model to handle such kind of data. And since it is tree based algorithm, normalization or standardization is not necessary in this case.

```python
# build the model
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=66)

# fit the model with train set
rf.fit(X_train, y_train)
```

```python
# Prediction on training set
y_train_predict = rf.predict(X_train)

# Calculate the accuracy score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

rf_acc = accuracy_score(y_train, y_train_predict)
rf_precision = precision_score(y_train, y_train_predict, average='weighted')
rf_recall = recall_score(y_train, y_train_predict, average='weighted')
rf_f1 = f1_score(y_train, y_train_predict, average='weighted')

print('RF accuracy: ', rf_acc)
print('RF precision: ', rf_precision)
print('RF recall: ', rf_recall)
print('RF f1: ', rf_f1)
```

```python
# Cross validation in case of overfitting
# Cross validation
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict

rf_cv_acc = cross_val_score(rf, X_train, y_train, cv=5, scoring="accuracy").mean() 
y_train_cv_pred = cross_val_predict(rf, X_train, y_train, cv=5)

rf_cv_precision = precision_score(y_train, y_train_cv_pred, average='weighted')
rf_cv_recall = recall_score(y_train, y_train_cv_pred, average='weighted')
rf_cv_f1 = f1_score(y_train, y_train_cv_pred, average='weighted')

print("rf CV accuracy: ", rf_cv_acc)
print("rf CV precision: ", rf_cv_precision)
print("rf CV recall: ", rf_cv_recall)
print("rf CV f1: ", rf_cv_f1)
```

The model is overfitting, hyperparameters need to be tuned.

```python
# use GridSearchCV to search for the best parameters
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'n_estimators': [20, 50, 100], 'max_depth': [5, 10, 15]} 
]

rf_grid_search = RandomForestClassifier(random_state=66)
grid_search = GridSearchCV(rf_grid_search, param_grid, cv=3,
                           scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_params_
```

```python
best_rf = grid_search.best_estimator_

# predict on train set
y_best_predict = best_rf.predict(X_train)

best_rf_acc = accuracy_score(y_train, y_best_predict)
best_rf_precision = precision_score(y_train, y_best_predict, average='weighted')
best_rf_recall = recall_score(y_train, y_best_predict, average='weighted')
best_rf_f1 = f1_score(y_train, y_best_predict, average='weighted')

print('best_rf accuracy: ', best_rf_acc)
print('best_rf precision: ', best_rf_precision)
print('best_rf recall: ', best_rf_recall)
print('best_rf f1: ', best_rf_f1)
```

```python
y_test_best_predict = best_rf.predict(X_test)

best_rf_acc_test = accuracy_score(y_test, y_test_best_predict)
best_rf_precision_test = precision_score(y_test, y_test_best_predict, average='weighted')
best_rf_recall_test = recall_score(y_test, y_test_best_predict, average='weighted')
best_rf_f1_test = f1_score(y_test, y_test_best_predict, average='weighted')

print('best_rf accuracy on test: ', best_rf_acc_test)
print('best_rf precision on test: ', best_rf_precision_test)
print('best_rf recall on test: ', best_rf_recall_test)
print('best_rf f1 on test: ', best_rf_f1_test)
```

Let's take a look at the feature importances:

```python
# plot the feature importances
features = X_train.columns
importances = best_rf.feature_importances_
indices = np.argsort(importances)

plt.figure(figsize=(10,8))
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```
